**Part I. Conceptual Understanding**

**1. Define non-monotonic reasoning in your own words.**
Non-monotonic reasoning is a kind of logic where your conclusion can change when you learn something new. It’s like changing your mind after getting more facts. This type of reasoning lets us correct our earlier beliefs. It’s flexible and more realistic because new information can always appear.

**2. How does non-monotonic reasoning differ from monotonic reasoning?**
Monotonic reasoning means that once something is proven true, it stays true forever, no matter what. Non-monotonic reasoning is different because new facts can change what we believe. It’s like first thinking something is right, but then realizing it’s wrong after finding out more. This shows how reasoning can adapt to new situations.

**3. Give a real-life situation where a conclusion must change after new information is added.**
I once thought my classmate was absent because I didn’t see them in school. Later, I found out they were attending an online meeting with the teacher. That new information made me change my conclusion. It shows how our reasoning can shift when we learn something new.

**4. What is a default rule? Provide one example.**
A default rule is something we assume is true unless there’s proof that it’s not. For example, we usually think that birds can fly. But when we find out about penguins, we realize there are exceptions. This shows how rules can change when we get new information.

**5. How do argumentation frameworks help AI systems decide between conflicting rules?**
Argumentation frameworks help AI choose between two conflicting ideas. They work by finding which rule is stronger or makes more sense in a given situation. The system can reject the weaker argument and follow the more logical one. This helps AI make smarter and more accurate decisions.

**Task 1: Belief Revision Simulation**

```python
def can_it_fly(animal):
    print(f"Checking if {animal} can fly...")

    birds = ["sparrow", "eagle", "penguin", "ostrich", "parrot"]
    exceptions = ["penguin", "ostrich"]

    if animal.lower() not in birds:
        print(f"{animal.capitalize()} is not a bird. So, the rule doesn’t apply.")
    else:
        print(f"{animal.capitalize()} is a bird. Default rule: Birds can fly.")
        if animal.lower() in exceptions:
            print(f"But {animal.capitalize()} is an exception to the rule.")
            print(f"Conclusion: {animal.capitalize()} cannot fly.")
        else:
            print(f"No exception found.")
            print(f"Conclusion: {animal.capitalize()} can fly.")

animal_name = input("Enter an animal name: ")
can_it_fly(animal_name)
```
